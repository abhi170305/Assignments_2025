```python
import os
from ultralytics import YOLO
```

```python
# Running the code on google colab
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Dataset stored as a zip file on drive
!unzip /content/drive/MyDrive/bullseye_data/augmented_data/combined_augmented_dataset.zip
```

```python
# Define paths
dataset_path = '/content/content/combined_augmented_dataset'
os.rename("/content/content/combined_augmented_dataset/valid", "/content/content/combined_augmented_dataset/val")
data_yaml = os.path.join(dataset_path, 'data.yaml')
```

```python
# Creating data.yaml file
with open("/content/content/combined_augmented_dataset/data.yaml", "w") as file:
    file.write(f"""path: {dataset_path}
train: train/images
val: val/images
test: test/images

nc: 1
names: ['0']"""
)
```

```python
# Load yolov8 small model
model = YOLO("yolov8s.pt")
```

```python
# Train the model
model.train(data='/content/content/combined_augmented_dataset/data.yaml',
            epochs=10,
            imgsz=320,
            batch=16,
            project='bullseye_runs',
            name='yolov8s_bullseye_10_320_16')
```

```python
# Create directory to store the runs on the drive
os.mkdir("/content/drive/MyDrive/bullseye_runs")
```

```python
import shutil

# Assuming 'bullseye_runs' exists in the current working directory
source_folder = '/content/bullseye_runs'
destination_folder = '/content/drive/MyDrive/bullseye_runs'

try:
    shutil.copytree(source_folder, destination_folder)
    print(f"Successfully copied '{source_folder}' to '{destination_folder}'")
except FileExistsError:
    print(f"'{destination_folder}' already exists.  Overwriting is not performed.")
except FileNotFoundError:
    print(f"'{source_folder}' not found. Please ensure the folder exists.")
except OSError as e:
    print(f"An error occurred: {e}")
```
